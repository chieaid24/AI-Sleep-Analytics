```python
import boto3
from botocore.exceptions import ClientError


bucket = 'sleep-quality-forecast-bucket'
key = 'SLEEP_RECORD.csv'
s3 = boto3.client('s3')
```

```python
import pandas as pd
from io import StringIO

# Load csv
obj = s3.get_object(Bucket=bucket, Key=key)
df = pd.read_csv(obj['Body'])
df.head()
```

```python
# Check missing values
print(df.isna().sum())

# Convert date columns
df['SESSION_DATE'] = pd.to_datetime(df['SESSION_DATE'])
```

```python
# select relevant columns
cols = ['SESSION_DATE', 'SLEEP_SCORE', 'USAGE_HOURS', 'AHI', 
        'MASK_SESSION_COUNT', 'LEAK_50_PERCENTILE', 'LEAK_70_PERCENTILE', 'LEAK_95_PERCENTILE']
df = df[cols].copy()

# sort by date
df = df.sort_values('SESSION_DATE')
# remove all days that Usage Hours = 0
df = df[df['USAGE_HOURS'] != 0]
```

```python
# Example: Day of week, rolling averages, usage change
df['DAY_OF_WEEK'] = df['SESSION_DATE'].dt.day_name()
```

```python
df['SESSION_DATE'].head()
print(df['SESSION_DATE'].dtype)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df['SESSION_DATE'], df['SLEEP_SCORE'], label='Sleep Score')
plt.plot(df['SESSION_DATE'], df['USAGE_HOURS'], label='Usage Hours')
plt.plot(df['SESSION_DATE'], df['AHI'], label='AHI')
plt.plot(df['SESSION_DATE'], df['MASK_SESSION_COUNT'], label='Mask')
plt.plot(df['SESSION_DATE'], df['LEAK_95_PERCENTILE'], label='Leak95')
plt.legend()
plt.title('Sleep Score and Usage Hours Over Time')
plt.show()

print(df.head(5))
```

```python
clean_key = 'processed/sleep_data_clean.csv'
csv_buffer = StringIO()
df.to_csv(csv_buffer, index=False)

s3.put_object(Bucket=bucket, Key=clean_key, Body=csv_buffer.getvalue())
print(f" Clean data saved to {bucket}/{clean_key}")
```
